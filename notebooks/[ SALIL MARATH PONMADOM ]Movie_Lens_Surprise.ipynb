### 1st PipeLine - Load data, Train Test split, model training, prediction, Evaluation.

 ### 1.1 Loading the Packages and Necesssary Data Set 

```python
import pandas as pd
import numpy as np 

from surprise import SVD
from surprise import Dataset
from surprise.model_selection import cross_validate
from surprise.model_selection import train_test_split
from surprise import Reader

from surprise import accuracy

import os
import sys
#data = pd.read_csv("ml-latest-small/ratings.csv")
#data.sample(5)
```

```python
# path to dataset file
file_path = os.path.expanduser('~/.surprise_data/ml-100k/ml-100k/u.data')

# As we're loading a custom dataset, we need to define a reader. In the
# movielens-100k dataset, each line has the following format:
# 'user item rating timestamp', separated by '\t' characters.
reader = Reader(line_format='user item rating timestamp', sep='\t')

data = Dataset.load_from_file(file_path, reader=reader)
```

```python
#data = Dataset.load_builtin('ml-100k')
#file_path = os.path.expanduser('~/.surprise_data/ml-100k/ml-100k/u.data')
#.....
```

### 1.2 Train - Test Split 

```python
trainset, testset = train_test_split(data, test_size=.25)
#.....
```

### 1.3 Model Training 

```python
# Use the famous SVD algorithm.
model = SVD()

# We use Item based approach 
```

### 1.4 Fit the Model and Predictions 

```python
# Train the algorithm on the trainset, and predict ratings for the testset
model.fit(trainset)
predictions = model.test(testset)
predictions
```

### 1.5 Evaluation 

```python
# Then compute RMSE
accuracy.rmse(predictions)
```

```python
accuracy.mae(predictions)
```

### 2nd Pipeline - Load the data and Crosss Validate 

### 2. Cross Validating on the Data

```python
# Cross Validation 
cross_validate(model, data , measures=['RMSE','MAE'],cv=5 , verbose = True)

# Here cv = 5 means it's the 5 random chosen groups of train and test splitting 
```

### 3. BenchMarking User based and Item Based 

### 3.1 Item Based 
### 3.1.1(Cosine Similarity metrices) 

```python
# Trying for Knn Algorithm  with cosine similarity 

from surprise import KNNBasic

sim_options = {'name': 'cosine',
               'user_based': False  # compute  similarities between items
               }
model_item_cos = KNNBasic(sim_options=sim_options)
```

#### Model fit and Predictions 

```python
model_item_cos.fit(trainset)

item_cos_pred = model_item_cos.test(testset)
item_cos_pred
```

#### Evaluation 

```python
display(accuracy.rmse(item_cos_pred))
display(accuracy.mae(item_cos_pred))
```

### 3.1.2 Pearson corelation Metrices 

```python
sim_options = {'name': 'pearson_baseline',  # here it uses the baselines centering rather than mean in pearson similarity 
               'shrinkage': 0  # no shrinkage
               }
model_item_pear = KNNBasic(sim_options=sim_options)
```

#### Model Fit and Prediction 

```python
model_item_pear.fit(trainset)

item_pear_pred = model_item_pear.test(testset)
item_pear_pred
```

#### Evaluation

```python
# Metrices 
display(accuracy.rmse(item_pear_pred))
display(accuracy.mae(item_pear_pred))
```
